# Requirements

```python
import cv2
import os
import matplotlib.pyplot as plt
```

# Loading Image

```python
# Define the path to the image file using os.path.join for cross-platform compatibility
image_address1 = os.path.join("images", "image.webp")
image_address2 = os.path.join("images", "image2.jpg")

# Define the path to the Haar Cascade model file using os.path.join for cross-platform compatibility
model = os.path.join("models", "haarcascade_frontalcatface_extended.xml")

# Read the input image from the specified path
image1 = cv2.imread(image_address1)
image2 = cv2.imread(image_address2)
# Convert the image from BGR to RGB format (as OpenCV uses BGR by default and Matplotlib uses RGB)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
# Create a figure with specified size to display the image
plt.figure(figsize=(10,7))
plt.imshow(image1)
plt.axis('off')
plt.figure(figsize=(10,7))
plt.imshow(image2)
plt.axis('off')
# Display the image using Matplotlib
#plt.imshow(image2) 
# Turn off the axis for a cleaner look
# Show the figure
plt.show()
```

# Detecting Cat

```python
# Load the Haar Cascade model for cat face detection
detector = cv2.CascadeClassifier(model)
if detector.empty():
    print("Error loading Haar Cascade XML file!")
else:
    print("Model loaded successfully!")

# Detect cat faces in the image using the loaded model
# scaleFactor: Parameter specifying how much the image size is reduced at each image scale
# minNeighbors: Parameter specifying how many neighbors each candidate rectangle should have to retain it
# minSize: Minimum possible object size. Objects smaller than this are ignored.

rects = detector.detectMultiScale(
    image1, 
    scaleFactor=1.1, 
    minNeighbors=3, 
    minSize=(200, 200)
)

print(rects)
# Iterate over the detected rectangles (bounding boxes for cat faces)
for (i, (x, y, w, h)) in enumerate(rects):
    # Draw a rectangle around the detected cat face
    cv2.rectangle(image1, (x, y), (x+w, y+h), (255, 0, 0), 2)
    # Annotate the image with the text 'Cat #N' above the rectangle
    text = f'Cat #{i+1}'
    cv2.putText(image1, text, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
    pass


rects = detector.detectMultiScale(
    image2, 
    scaleFactor=1.1, 
    minNeighbors=1, 
    minSize=(115, 115)
)

print(rects)
# Iterate over the detected rectangles (bounding boxes for cat faces)
for (i, (x, y, w, h)) in enumerate(rects):
    # Draw a rectangle around the detected cat face
    cv2.rectangle(image2, (x, y), (x+w, y+h), (255, 0, 0), 2)
    # Annotate the image with the text 'Cat #N' above the rectangle
    text = f'Cat #{i+1}'
    cv2.putText(image2, text, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)
    pass
```

```python
# Create a figure with specified size to display the image
plt.figure(figsize=(10,7))
plt.imshow(image1)
plt.axis('off')

# Display the image with detected cat faces and annotations
 
plt.figure(figsize=(10,7))
plt.imshow(image2)
plt.axis('off')
# Turn off the axis for a cleaner look


# Show the figure
plt.show()
```

```python
image_rgb1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image_rgb2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
cv2.imwrite('outputs/result1.jpg', image_rgb1)
cv2.imwrite('outputs/result2.jpg', image_rgb2)
```
